```python
import pandas as pd
import numpy as np
```

```python
Ind2SkillMat = pd.read_csv("Forecast/Industry to Skill Matrix.csv")
Ind2SkillMat = Ind2SkillMat.rename(columns={'Unnamed: 0': 'Industry'})
Ind2SkillMat = Ind2SkillMat.set_index('Industry')
Ind2SkillMat
```

```python
IndForecastMat = pd.read_csv("Forecast/Industry Forecast.csv")
IndForecastMat
```

```python
print(Ind2SkillMat.index)
print(IndForecastMat.columns)
```

```python
IndForecastMat = IndForecastMat.drop(['Transport.Equipment', 'Construction', 'Wholesale.Trade', 'Retail.Trade', 
                     'Water.Transport...Supporting.Services', 'Other.Transportation...Storage.Services', 
                    'Accommodation', 'Food...Beverage.Services', 'Real.Estate.Services', 'Other.Professional.Services',
                    'Security...Investigation', 'Cleaning...Landscaping'], axis=1)
```

```python
IndForecastMat.columns = ["Time"] + list(Ind2SkillMat.index)
IndForecastMat = IndForecastMat.set_index("Time")
IndForecastMat
```

```python
SkillForecastMat = pd.DataFrame(0, index=list(IndForecastMat.index), columns=list(Ind2SkillMat.columns)[1:])
SkillForecastMat
```

```python
SkillForecastMat = IndForecastMat.dot(Ind2SkillMat)
SkillForecastMat
```

```python
SkillForecastMat.to_csv("")
```
